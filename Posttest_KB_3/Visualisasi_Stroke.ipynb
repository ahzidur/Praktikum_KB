```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid") # Plih Style

df = pd.read_csv("Stroke.csv")
```

```python
df.head()
```

```python
#Grafik Batang
plt.figure(figsize=(10,10)) # panjang 10, lebar 10
sns.countplot(data=df,x= 'smoking_status', palette='crest', hue='gender')
plt.title('Perbandingan Status Merokok Tiap Gender')
plt.xlabel('Status Merokok')
plt.ylabel('Kuantitas')
plt.grid()
plt.show()
```

#### Grafik Batang yang menggunakan seaborn yang menjelaskan tentang jumlah data status merokok setiap pada jenis kelamin orang masing-masing dari dataset yang berjumlah 3000 data dengan xlabel: Status Merokok dan ylabel: Kuantitas

```python
plt.figure(figsize=(10,10))
sns.countplot(data=df,x= 'gender',palette='flare')
plt.title("Perbandingan Jenis Kelamin Pada Data")
plt.xlabel('Jenis Kelamin')
plt.ylabel('Kuantitas')
plt.show()
```

#### Grafik Batang yang menggunakan seaborn yang menjelaskan tentang jumlah data jenis kelamin pada dataset yang berjumlah 3000 data dengan xlabel: Jenis Kelamin dan ylabel: Kuantitas

```python
#Grafik Pie
data_pie = df["gender"].value_counts().rename_axis("gender").reset_index(name="id_count")
```

```python
plt.figure(figsize=(8,8))
plt.pie(data_pie["id_count"],labels=data_pie["gender"],startangle=120,autopct='%.1f%%')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Kuantitas')
plt.show()
```

#### Grafik Pie ini menampilkan data dalam bentuk Persenan dari Jumlah Jenis Kelamin yang berada pada dataset yang totalnya berjumlah 3000

```python
#Grafik HeatMap
fig, ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(), center=0, cmap='Blues')
ax.set_title('Korelasi antar atribut pada stroke')
```

#### Grafik HeatMap ini merupakan visualisasi atau pemetaan dengan menampilkan dari dataset stroke pada tiap jenis kelamin, yang menjelaskan korelasi antar variabel yang semakin tinggi angka suatu kelompok data maka warnanya akan semakin gelap.
